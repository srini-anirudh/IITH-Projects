```python
from skmeans import SKMeans
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from itertools import permutations
```

```python
# Dataset
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=1000, n_features=2, centers=3, cluster_std=2.0, random_state=42)
```

```python
#Scatter Plot
random_state = 42

# Taking different value of alpha and beta
alpha, beta = np.meshgrid(np.arange(0, 5, 1), np.arange(0, 5, 1))
alpha, beta = alpha.flatten(), beta.flatten()
```

```python
k = 3
plt.figure(figsize=(25, 25))
idx = 1

for a, b in zip(alpha, beta):
    kmeans = kmeans = SKMeans(no_clusters=k, 
                              iters=500, 
                              metric="linear_combination",
                              alpha=a, 
                              beta=b, 
                              random_state=random_state)
```
